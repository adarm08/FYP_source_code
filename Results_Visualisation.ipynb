```python
import numpy as np
import matplotlib.pyplot as plt
# Technical Models Test accuries
# Sample data
model1 = np.array([51, 48.6, 53.1, 51.5, 49.3, 51.3])
model2 = np.array([52.2, 44.8, 53.1, 49, 47.2, 48.9])
model3 = np.array([51.1, 48, 53.2, 51.2, 53.1, 51.3])

# Set up plot
fig, ax = plt.subplots(figsize=(6, 4))
index = np.arange(6)
bar_width = 0.2
opacity = 0.9

# Plot bars for each model
rects1 = ax.bar(index, model1, bar_width,
                alpha=opacity,
                color='#1f77b4',
                label='Model 1')
rects2 = ax.bar(index + bar_width, model2, bar_width,
                alpha=opacity,
                color='#ff7f0e',
                label='Model 2')
 
rects3 = ax.bar(index + 2*bar_width, model3, bar_width,
                alpha=opacity,
                color='#2ca02c',
                label='Model 3')

ax.set_ylim(top=60)

# Add data labels
for rect in rects1 + rects2 + rects3:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=8, rotation=90)

# Add axis labels and legend
ax.set_xlabel('Algorithm')
ax.set_ylabel('Accuracy %')
ax.set_title('Test Accuracy For Technical Models')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(('LR', 'RF', 'SVM', 'ANN', 'GRU', 'LSTM'))

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

plt.tight_layout()
plt.show()
```

```python
models = [model1, model2, model3]

for i, model in enumerate(models):
    avg = np.mean(model)
    print(f"Average for model {i+1}: {avg}")
```

```python
# News Based model

# Sample data
model1 = np.array([51.3, 52.2, 53.1, 53, 51.9, 53.7])
model2 = np.array([51.5, 51.3, 53, 51.9, 52.8, 53.3])
model3 = np.array([51.4, 48.9, 53.1, 52.7, 52.7, 53.6])
model4 = np.array([51.4, 52.2, 53.1, 53.4, 54.2, 56])
model5 = np.array([51, 51, 53.1, 52.3, 53.8, 52.1])
model6 = np.array([52.8, 51.8, 53.1, 53.9, 50.4, 54.7])
model7 = np.array([53, 52.5, 53, 50.1, 54, 50.8])

# Set up plot
fig, ax = plt.subplots(figsize=(11, 5))
index = np.arange(6)
bar_width = 0.13
opacity = 0.9

# Plot bars for each model
rects1 = ax.bar(index, model1, bar_width,
                alpha=opacity,
                color='#1f77b4',
                label='tf.idf (df > 3)')
rects2 = ax.bar(index + bar_width, model2, bar_width,
                alpha=opacity,
                color='#ff7f0e',
                label='tf.idf (df > 50)')
rects3 = ax.bar(index + 2*bar_width, model3, bar_width,
                alpha=opacity,
                color='#2ca02c',
                label='tf.idf (df > 100)')
rects4 = ax.bar(index + 3*bar_width, model4, bar_width,
                alpha=opacity,
                color='#d62728',
                label='tf.idf (df > 200)')
rects5 = ax.bar(index + 4*bar_width, model5, bar_width,
                alpha=opacity,
                color='purple',
                label='tf.idf (df > 100 + sentiment score)')
rects6 = ax.bar(index + 5*bar_width, model6, bar_width,
                alpha=opacity,
                color='gray',
                label='tf.idf (sentiment dictionary)')
rects7 = ax.bar(index + 6*bar_width, model7, bar_width,
                alpha=opacity,
                color='brown',
                label='GloVe')
ax.set_ylim(top=65)
# Add data labels
for rect in rects1 + rects2 + rects3 + rects4 + rects5 + rects6 + rects7:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=8, rotation=90)

# Add axis labels and legend
ax.set_xlabel('Algorithm')
ax.set_ylabel('Accuracy %')
ax.set_title('Test Accuracy For News Based Models')
ax.set_xticks(index + 3*bar_width)
ax.set_xticklabels(('LR', 'RF', 'SVM', 'ANN', 'GRU', 'LSTM'))

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=4)

plt.tight_layout()
plt.show()
```

```python
# Sample data
stocks = ['JPM', 'MSFT', 'KO', 'JNJ']
val_acc = [57.5, 58.7, 55.6, 56]
test_acc = [54.8, 51.6, 45.6, 53.2]

# Set up plot
fig, ax = plt.subplots(figsize=(8, 5))
index = np.arange(len(stocks))
bar_width = 0.4
opacity = 0.9

# Plot bars for validation accuracy
rects1 = ax.bar(index, val_acc, bar_width,
                alpha=opacity,
                color='#ff7f0e',
                label='Validation')

# Plot bars for test accuracy
rects2 = ax.bar(index + bar_width, test_acc, bar_width,
                alpha=opacity,
                color='#2ca02c',
                label='Test')

# Add data labels
for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate('{}'.format(height),
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom', fontsize=8, rotation=0)

# Add axis labels and legend
ax.set_ylim(top=80)
ax.set_xlabel('Stock Ticker')
ax.set_ylabel('Accuracy %')
ax.set_title('Validation and Test Accuracy for Individual Stocks')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(stocks)
ax.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
